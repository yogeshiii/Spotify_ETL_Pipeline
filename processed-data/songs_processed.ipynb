```python
import pandas as pd
import numpy as np
```

```python
df_song = pd.read_csv("songs.csv")
df_song.head()
```

```python
df_song.columns
```

```python
df_song = df_song.drop(
    ['danceability', 'energy', 'key',
       'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness',
       'liveness', 'valence', 'tempo', 'duration_ms', 'lyrics', 'year', 'popularity', 'total_artist_followers',
       'avg_artist_popularity', 'niche_genres'],
    axis=1
)
```

```python
df_song.head()
```

```python
df_song["name"] = df_song["name"].astype(str)
df_song["name"] = df_song["name"].str.replace(r'[!"]|\.','',regex=True)
```

```python
df_song["name"] = df_song["name"].str.lower()
```

```python
df_song["album_name"] = df_song["album_name"].astype(str)
df_song["album_name"] = (df_song["album_name"]
                        .str.replace(r"+", "", regex = False)
                        .str.replace(r"'", "", regex = False)
                        .str.replace(r".", "", regex = False)
                        .str.replace(r'"', '', regex = False)
                        .str.replace(r'!', '', regex = False))
df_song["album_name"] = df_song["album_name"].str.lower()
```

```python
df_song["artists"] = df_song["artists"].astype(str)
df_song["artists"] = df_song["artists"].str.replace(r'[\[\]"]', '', regex=True)
df_song["artists"]
```

```python
df_song["artist_ids"] = df_song["artist_ids"].astype(str)
df_song["artist_ids"] = df_song["artist_ids"].str.replace(r'[\[\]"]', '', regex=True)
df_song["artist_ids"]
```

```python
df_song = df_song[
    df_song["name"].notna() &
    (df_song["name"].str.strip() != "") & (df_song["name"].str.strip() != " ")
]
```

```python
df_song["artists"] = df_song["artists"].str.lower()
```

```python
df_song["genre"] = df_song["genre"].str.lower()
```

```python
df_song = df_song.drop_duplicates(subset=["name"], keep="first")
```

```python
df_song = df_song.dropna(subset=["name"])
```
